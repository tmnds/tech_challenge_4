```python
import requests
import json
```

```python
body = {
  "end_date": "2024-11-15",
  "start_date": "2024-06-01",
  "seq_length": 20,
  "horizon": 1
}
```

```python
res = requests.post(url="http://127.0.0.1:5000/predict",data=json.dumps(body))
```

```python
result = json.loads(res.text)
dates_raw = result["input df"]["Datetime"]
y_pred_raw = result["prediction"]
y_test_raw = result["input df"]["PETR4.SA"]
```

```python
result
```

```python
y_pred_raw
```

```python
start_index = int(list(y_pred_raw.keys())[0])
end_index = int(list(y_test_raw.keys())[-1])
print([start_index, end_index])
```

```python
y_pred = []
y_test = []
for k in range(start_index,end_index+1):
    y_pred.append(y_pred_raw[str(k)])
    y_test.append(y_test_raw[str(k)])
```

```python
y_pred
```

```python
import matplotlib.pyplot as plt

# Plot the predicted prices\n"
plt.plot(y_pred, label="Predicted Prices")
plt.plot(y_test, label="Real Prices")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import root_mean_squared_error, r2_score

# Calculate RMSE and R score
rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print([ f"R2={r2}", f"RMSE={rmse}"])
```
